```python
import numpy as np

# super -> (x,0,0)
# super with brackets -> (x,1,0)
# sub -> (x,0,1)
# sub with brackets -> (x,1,1)

def getscrpt(x,flag,subs=0):
    if(type(x)==int):
        x = str(x)
    normal = "t0123456789+-=()"
    super_s = "ᵗ⁰¹²³⁴⁵⁶⁷⁸⁹⁺⁻⁼⁽⁾"
    sub_s = "ₜ₀₁₂₃₄₅₆₇₈₉₊₋₌₍₎"
    if subs==1:
        res = x.maketrans(''.join(normal), ''.join(sub_s))
    else:
        res = x.maketrans(''.join(normal), ''.join(super_s))
    if flag==1:
        if subs==1:
            return "₍"+x.translate(res)+"₎"
        else:
            return "⁽"+x.translate(res)+"⁾"
    else:
        return x.translate(res) 
```

```python
A = np.array([[1,2,3,4],
             [4,5,6,7],
             [2,1,5,0],
             [4,2,1,0]])
```

# ITERATION 1

```python
# Edit d & e here
d = np.array([[4],[2],[4]]) 
e = np.array([[1],[0],[0]])
# ---------------------------------------

Beta = -1 * np.linalg.norm(d,2)
print("Beta = -||d||{} = {}".format(getscrpt(2,0,1),Beta))

dBe = d-Beta*e
print("d-Be{} = {}{}".format(getscrpt(1,1,0),np.transpose(dBe)[0],getscrpt("t",0,0)))

dBeNorm = np.linalg.norm(dBe,2)
print("||d-Be{}||{} = {:.5f}".format(getscrpt(1,1,0),getscrpt(2,0,1),dBeNorm))

Alpha = 1.414/dBeNorm
print("Alpha = {:.5f}".format(Alpha))

v = Alpha*dBe
print("v = Alpha*[d-Be{}] = {}{}".format(getscrpt(1,1,0),np.transpose(v)[0],getscrpt("t",0,0)))

vvt = np.dot(v,np.transpose(v))
I = np.identity(vvt.shape[0])
U = I-vvt
print("\nvv*=",vvt)
print("\nU =",U)

U = np.pad(U,((1, 0), (1, 0)))
U[0][0]=1
print("\nPadded U =",U)

H1 = np.dot(U,np.dot(A,U))
np.set_printoptions(formatter={'float_kind':'{:f}'.format})
print("\nH{} = UAU* = {}".format(getscrpt(1,0,1),H1))        
```
